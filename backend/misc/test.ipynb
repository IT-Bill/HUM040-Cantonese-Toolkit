```python
import base64
from tencentcloud.common import credential
from tencentcloud.common.exception.tencent_cloud_sdk_exception import TencentCloudSDKException
from tencentcloud.asr.v20190614 import asr_client, models

# 初始化凭证
cred = credential.Credential(
    # remove ID
)
client = asr_client.AsrClient(cred, "ap-guangzhou")

# 准备请求
req = models.SentenceRecognitionRequest()
req.EngSerViceType = "16k_yue"
req.headers = {"Content-Type": "application/json; charset=utf-8"}
req.VoiceFormat = "mp3"
req.SourceType = 1

# 读取音频文件
with open("voice/common_voice_yue_37290287.mp3", "rb") as f:
    audio_data = f.read()
req.Data = base64.b64encode(audio_data).decode()

response = client.SentenceRecognition(req)
print(response.to_json_string())
```

```python
import wave

# PCM 文件的参数
pcm_file = 'demo.pcm'  # PCM 文件的路径
wav_file = 'output.wav'  # 输出的 WAV 文件的路径
sample_rate = 8000  # 采样率
channels = 2  # 声道数
sample_width = 2  # 采样字节宽度，2 代表 16 位

# 读取 PCM 数据
with open(pcm_file, 'rb') as pcmfile:
    pcm_data = pcmfile.read()

# 使用 wave 模块写入 WAV 文件
with wave.open(wav_file, 'wb') as wavfile:
    # 设置参数
    wavfile.setframerate(sample_rate)
    wavfile.setnchannels(channels)
    wavfile.setsampwidth(sample_width)
    # 写入数据
    wavfile.writeframes(pcm_data)

# 这里我们没有返回语句，因为这是一个独立的脚本
# 运行此脚本会将指定的 PCM 文件转换为 WAV 文件
# 请确保提供的 pcm_file 变量包含正确路径的 PCM 文件名
# 并且有相应的读取权限

```
